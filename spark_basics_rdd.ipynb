```python
from pyspark import SparkContext, SparkConf
```

### Creating a SparkContext instance

```python
sc = SparkContext.getOrCreate()
```

### creating an RDD (Resilient Distributed Dataset - a collection of elements that can be operated on in parallel)

```python
data = [1,2,3,4,1,2,3,4,1,4,7,6,8,2,4,2,1]
rdd = sc.parallelize(data)
```
